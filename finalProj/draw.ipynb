```python
import matplotlib.pyplot as plt
import re

# 读取日志文件
log_file = 'loss_log.txt'

# 初始化损失函数列表
D_A_vals = []
G_A_vals = []
cycle_A_vals = []
idt_A_vals = []
D_B_vals = []
G_B_vals = []
cycle_B_vals = []
idt_B_vals = []

# 正则表达式用于匹配日志中的数值
log_pattern = r"D_A: ([\d.]+) G_A: ([\d.]+) cycle_A: ([\d.]+) idt_A: ([\d.]+) D_B: ([\d.]+) G_B: ([\d.]+) cycle_B: ([\d.]+) idt_B: ([\d.]+)"

# 读取文件并提取数据
with open(log_file, 'r') as file:
    for line in file:
        match = re.search(log_pattern, line)
        if match:
            D_A_vals.append(float(match.group(1)))
            G_A_vals.append(float(match.group(2)))
            cycle_A_vals.append(float(match.group(3)))
            idt_A_vals.append(float(match.group(4)))
            D_B_vals.append(float(match.group(5)))
            G_B_vals.append(float(match.group(6)))
            cycle_B_vals.append(float(match.group(7)))
            idt_B_vals.append(float(match.group(8)))

# 隔一个画一个
D_A_vals = D_A_vals[::2]
G_A_vals = G_A_vals[::2]
cycle_A_vals = cycle_A_vals[::2]
idt_A_vals = idt_A_vals[::2]
D_B_vals = D_B_vals[::2]
G_B_vals = G_B_vals[::2]
cycle_B_vals = cycle_B_vals[::2]
idt_B_vals = idt_B_vals[::2]

# 绘制损失函数图像
epochs = range(1, len(D_A_vals) + 1)

plt.figure(figsize=(32, 16))



# 绘制D_A, G_A, cycle_A, idt_A的变化图
plt.subplot(2, 2, 1)
plt.plot(epochs, D_A_vals, label='D_A', color='blue')
plt.plot(epochs, G_A_vals, label='G_A', color='orange')
# plt.plot(epochs, D_A_vals, label='D_A', color='blue')
# plt.plot(epochs, G_A_vals, label='G_A', color='orange')
# plt.plot(epochs, cycle_A_vals, label='cycle_A', color='green')
# plt.plot(epochs, idt_A_vals, label='idt_A', color='red')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('G_A & D_A Losses')
plt.legend()

# 绘制D_B, G_B, cycle_B, idt_B的变化图
plt.subplot(2, 2, 3)
plt.plot(epochs, cycle_A_vals, label='cycle_A', color='green')
plt.plot(epochs, idt_A_vals, label='idt_A', color='red')
# plt.plot(epochs, D_B_vals, label='D_B', color='blue')

# plt.plot(epochs, G_B_vals, label='G_B', color='orange')
# plt.plot(epochs, cycle_B_vals, label='cycle_B', color='green')
# plt.plot(epochs, idt_B_vals, label='idt_B', color='red')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('cycle_A & idt_A Losses')
plt.legend()

plt.tight_layout()
plt.show()
```
